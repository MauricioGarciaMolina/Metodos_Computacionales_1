```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
```

$$ \vec{r} (t) = Acos(\omega t)\hat{i} +   Asin (\omega t)\hat{j} $$

```python
A=1
omega=2*np.pi/10
N=int(1e2)
```

```python
t=np.linspace(0,10,N)
r=np.zeros((N,2))
```

```python
def GetPosition():
    r[:,0] =A*np.cos(omega*t)
    r[:,1] =A*np.sin(omega*t)
```

```python
GetPosition()
```

```python
#plt.scatter(r[:,0],r[:,1],color="black",s=0.5)
#plt.axis("equal")
```

```python
fig = plt.figure( figsize=(4,2))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

def init():
    
    ax2.set_ylim(-A,A)
    ax1.set_ylim(-A,A)

def Update(i):
    
    plot = ax1.clear()
    plot = ax2.clear()
    
    init()
    
    plot = ax1.plot(t[:i],r[:i,0],color="b")
    plot = ax2.plot(t[:i],r[:i,1],color="k")
    return plot

Anim = anim.FuncAnimation(fig, Update, frames=N, init_func=init )
```

```python
v=np.zeros((N,2))
```

```python
def GetVelocity():
    v[:,0]=-A*omega*np.sin(omega*t)
    v[:,1]=A*omega*np.cos(omega*t)
```

```python
GetVelocity()
```
