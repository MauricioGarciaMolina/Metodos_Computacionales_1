Newton Rapson:

Como en la curva, f es continua $\to$ f es derivable $\therefore \exists Df$

$Df(x_n)=\dfrac{f(x_{n+1}-f{x_n}}{x_{n+1}-x_n}$

Error $O(h^2)$

Forma iterativa:

$x_{n+1}=x_n-\dfrac{f(x_n)}{f'(x_n)}$

Criterio de Parada:

$e=\dfrac()$

```python
import numpy as np
```

```python
def derivada_central(f,x,h):
    d_f= (f(x+h)-f(x-h))/(2*h)
    return d_f
```

```python
def Funtion(x):
    return 5*(1-np.exp(-x))-x
```

```python
def NewtonRapson(f,x_0,h):
    precision=h
    x_p=x_0
    error=1
    it=0
    itmax=int(1/h)
    while(error>precision and it<itmax):
        try:
            x_n = x_p - (f(x_p)/derivada_central(f,x_p,h))
            error=np.abs(f(x_n))/np.abs(derivada_central(f,x_n,h))
        except ZeroDivisionError:
            print("División por cero.")
        x_p=x_n
        it+=1
    print(f"Raíz: {x_n},Número de pasos = {it}")
    if(it==itmax):
        return False
    else:
        return x_n
```

```python
def raices(f,x,h,tol=8):
    roots=np.array([])
    for i in x:
        root=NewtonRapson(f,i,h)
        if(root!=False):
            croot=np.round(root, tol)
            if(croot not in roots):
                np.append(roots,croot)
    roots.sort()
    return roots
```
