Clase Introductoria Sympy

Sympy es una libreria para hacer Algebra dentro de python.

Es compatible con Numpy

```python
import sympy as sym
import numpy as np
import matplotlib as rc
import matplotlib.pyplot as plt
```

```python
x=sym.Symbol("x",Real=True)
func=sym.exp(-0.5*x)*sym.sin(5*x)
```

```python
func
```

```python
f= sym.lambdify([x],func,"numpy")
```

```python
f(0.3)
```

```python
x=np.linspace(0,2*np.pi,int(1e5))
plt.plot(x,f(x))
```

Legendre:

```python
def GetLegendre(n):
    x=sym.Symbol("x",Real=True)
    y=sym.Symbol("y",Real=True)
    
    y=(x**2-1)**n
    
    p = sym.diff(y,x,n)/(2**n * np.math.factorial(n))
    
    return p
```

```python
GetLegendre(0)
```

```python
Legendre=[]
n=10
for i in range(n+1):
    termino_n=GetLegendre(i)
    Legendre.append(termino_n)
```

```python
print(Legendre)
```

```python
fig = plt.figure(figsize=(5,5))
ax=fig.add_subplot(1,1,1)
t=np.linspace(-1,1,int(1e5))
x=sym.Symbol("x",Real=True)
for i in range(1,n+1):
    f= sym.lambdify([x],Legendre[i],"numpy")
    ax.plot(t,f(t))
```

```python
print(f)
```

```python
def GetLaguerre(n):
    x=sym.Symbol("x",Real=True)
    y=sym.Symbol("y",Real=True)
    
    y= sym.exp(-x)*x**n
    
    p = ((sym.exp(x))/(np.math.factorial(n)))*sym.diff(x,y,n)
    
    return p
```

```python
GetLaguerre(1)
```
